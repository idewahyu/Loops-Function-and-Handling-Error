##Function
In Python, a function is a group of related statements that performs a specific task. Functions help break our program into smaller and modular chunks. As our program grows larger and larger, functions make it more organized and manageable.

Furthermore, it avoids repetition and makes the code reusable.

##Defining Function

- Functions are defined with the def keyword followed by the name of the function and its parameters in brackets ().
- Optionally, you can add a docstring (a documentation string that describes the context of the function).
- The code block in each function starts with a colon and uses an indent.
- The function stops when there is a return [expression] statement that returns [expression].
- You can also make the function return no output by returning None.

```python
def square(x):
    square2 = x**2 #fungsi pangkat dua
    return square2
```

```python
square(5)
```

```python
def name(x):
    hello = "Hi {}".format(x)
    return hello
```

```python
name("John")
```

#Return
The return statement is used to exit a function and go back to the place from where it was called.

```python
def addition(x,y):
    plus = x + y #fungsi penjumlahan
    print("Result of this function is {}".format(plus))
    return plus
```

```python
addition(2,3)
```

```python
output = addition(2,3)
print("Result of this return is {}".format(output))
```

- The return value of a function can be stored in a variable.
- This will distinguish a function that returns a value from a function that does not return a value (procedure)

```python
def square(x):
    square3 = x**2
    return square3
```

```python
square(5)
```

```python
number = int(input().strip())
result = square(number)
print("The result of this function is {}".format(result))
```

```python
print("{} squared is {}".format(number, result))
```

#Pass by Reference
By default, python pass the variable to a function as a reference. In this technique, the memory location of the variable is sent to the function.

If the function is able to change the original value of the variable using the reference. We call it passing-by-reference.

```python
def list(list_number):
    list_number.append([3,4])
    print("The result of this function is {}".format(list_number))
```

```python
list_number = [5,6]
list(list_number)
print("The value from outside the function is {}".format(list_number))
```

However, we must be careful because setting a variable with the same name as a parameter, means creating a new variable in the local scope and not related to global variables.

```python
def list(list_number):
    list_number = [3,4]
    print("The result of this function is {}".format(list_number))
```

```python
list_number = [5,6]
list(list_number)
print("The result of this function is {}".format(list_number))
```
