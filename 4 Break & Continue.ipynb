#Break and Continue
Break is used to exit a for loop or a while loop, whereas continue is used to skip the current block, and return to the "for" or "while" statement.

1. Break
 
 - Break
 - Break stops the loop then exits, followed by executing the statement after the loop block.
 - If you have a nested loop, break will stop the loop according to the level or loop it is in.
 - But if it is placed in a loop with a second depth for example, only that loop will stop, not with the main loop.

```python
for i in "Data Science":
    if i == " ": 
      break
    print(i)
```

```python
for i in "Python for Data Science":
    if i == "D":
      break
    print(i)
```

2. Continue

  - The continue statement will stop the current iteration.
  - Then move on to the next iteration.
  - Ignore statements that are between continue and the end of the loop block.

```python
for i in "Data":
    if i == "a": 
        continue
    print(i)
```
