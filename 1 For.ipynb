#LOOPS
There are two types of loops in Python: For and While.

##1. For
  - A for loop is used for iterating over a sequence (that is either a list, a tuple, a dictionary, a set, or a string).
  - If a list or sequence contains an expression, it will be evacuated first.
  - Then the first item in the sequence/list will be assigned as iterating variabel.
  - After that, the statement block will be executed, continue to the next item until the entire sequence is exhausted.


```python
for i in 'data':
  print(i)
```

```python
for i in 'data':
    print("letter {}".format(i))
```

```python
colors = ['red', 'yellow', 'green']

for i in colors:
    print("color {}".format(i))
```

```python
for i in range(3):
    print(i)
```

```python
for i in range(0,3):
    print(i)
```

###Nested Loop
- A nested loop is a loop inside the body of the outer loop. The inner or outer loop can be any type, such as a while loop or for loop. 
The outer loop can contain more than one inner loop. There is no limitation on the chaining of loops.

- In the nested loop, the number of iterations will be equal to the number of iterations in the outer loop multiplied by the iterations in the inner loop.

- In each iteration of the outer loop inner loop execute all its iteration. For each iteration of an outer loop the inner loop re-start and completes its execution before the outer loop can continue to its next iteration.

- Nested loops are typically used for working with multidimensional data structures, such as printing two-dimensional arrays, iterating a list that contains a nested list.

###Here an example of Python Nested: "FOR LOOP"



```python
for row in range(3): #boolean logic
    for col in range(3): #optional
        if row == 0 and col <= 2: 
            print("*", end = " ") 
        elif row == 1 and col <=1: 
            print("*", end = " ") 
        elif row == 2 and col == 0: 
             print("*", end = " ")
    print() 
```

##The Else Statement After For Loop

The else function after for is a function that takes precedence over search loops - to provide a program exit when the search is not found.

```python
for i in range(1,7):
    if i % 2 == 0:
        print("{} is an even number".format(i))
        break
    else:
        print("{} is not an even number".format(i))
```
