#While
- While loops repeat as long as a certain boolean condition is met (true).
- If temporary, the condition is limited.
- The condition can be any expression, and keep in mind that True in Python includes all non-zero values.
- When the condition is False, the program will continue to the line after the statement block.

```python
angka = int(input())

while (angka < 3):
  print("bilangan")

  #If while is not given a limit it will continue to print and will not stop.
  #While must be given a limiting condition!!!
```

```python
angka = int(input()) 

while (angka < 3):
    print("bilangan {}".format(angka))
    angka = angka + 1 
```
